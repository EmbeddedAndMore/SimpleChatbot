# 1-import doc2vec packages and use it for sentence similarity

```python
import gensim
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from gensim.utils import simple_preprocess
import multiprocessing
import os
```

#### prepare generator object to pre process text a little bit

```python
class MyTexts(object):
    def __iter__(self):
        for i in range(len(df2)):
            yield TaggedDocument(words=simple_preprocess(df2['text'][i]), tags=[i])
```

#### train model and save it at the end.

```python
%%time

cores = multiprocessing.cpu_count()
print(f'you have {cores}-core CPU')

texts = MyTexts()
doc2vec_model = Doc2Vec(size=350, workers=cores)
doc2vec_model.build_vocab(texts)
doc2vec_model.train(texts, total_examples=doc2vec_model.corpus_count, epochs=15)

if not os.path.exists('models'):
    os.makedirs('models')
    doc2vec_model.save('models/doc2vec.model')
else:
    doc2vec_model.save('models/doc2vec.model')
print('And we did it!')
```

#### a little bit of sanity check

```python
print(doc2vec_model.most_similar(['hey'])[0:2])
print(doc2vec_model.most_similar(['google'])[0:2])
```

#### Chatbot class to get input from user and give it to the model to find similar questions

```python
def chatbot():
    quit = False
    while quit == False:
        text = str(input('Me: '))
        text = text.lower()
        ##an optional quit command
        if text == 'quit()':
            quit=True
        else:
            tokens = text.split()
            ##infer vector for text the model may not have seen
            new_vector = doc2vec_model.infer_vector(tokens)
            ##find the most similar [i] tags
            index = doc2vec_model.docvecs.most_similar([new_vector], topn = 10)
            print(index)
            print('Chatbot: ' + df2.iloc[index[0][0],0])
            print('Chatbot: ' + df2.iloc[index[0][0],1])
            print('\n')
```

#### Check weather API

```python
def get_city_code(reponse) -> (str, str):
    if reponse.status_code == 200:
        js = json.loads(reponse.content)
        if len(js) > 0:
            key = js[0]["Key"]
            name = js[0]["EnglishName"]
            return key, name
        return None, None
```

```python
def get_city_weather(reponse) -> (int, int, str, str):
    if reponse.status_code == 200:
        js = json.loads(reponse.content)
        minimum = js['DailyForecasts'][0]['Temperature']['Minimum']['Value']
        maximum = js['DailyForecasts'][0]['Temperature']['Maximum']['Value']
        day_stat = js['DailyForecasts'][0]['Day']['IconPhrase']
        night_stat = js['DailyForecasts'][0]['Night']['IconPhrase']
        return minimum, maximum, day_stat, night_stat
    return None, None, None, None
```

#### Test DockDockGo search

```python
def search_dock(search_phrase)-> str:
    encoded_search_phrase = urllib.parse.quote(search_phrase)
    url ="https://api.duckduckgo.com/?q="+ encoded_search_phrase + "&format=json&pretty=10"
    dock_resp = requests.get(url)
    if dock_resp.status_code == 200:
        js = json.loads(dock_resp.content)
        summary = js["Abstract"]
        if len(summary.strip()) > 0:
            return summary.strip()
    return None
    
```

#### create chat bot by finding closest sentence using Cosine similarity

```python
def chatbot1():
    quit = False
    while quit == False:
        text = str(input('Me: '))
        queries = [text]
        ##an optional quit command
        if text == 'quit()':
            quit=True
        else:
            if "weather" in text and "forecast" in text:
                city = str(input('Please write the city name again: '))
                params = {
                    "apikey":weather_api_key,
                    "q":city,
                    "language":"en-us"
                }
                r_city = requests.get('http://dataservice.accuweather.com/locations/v1/cities/search', params=params)
                if r_city.status_code != 200 :
                    print("City not found!")
                else:
                    c_key, c_name = get_city_code(r_city)
                    if c_key is not None:
                        print(f"City {c_name} found!")
                        params = {
                            "apikey":weather_api_key,
                            "language":"en-us",
                            "metric": True
                        }
                        w_resp = requests.get('http://dataservice.accuweather.com/forecasts/v1/daily/1day/'+ c_key, params=params)
                        mini, maxi, day, night = get_city_weather(w_resp)
                        print(f"Min Temp: {mini}\nMax Temp: {maxi}\nDay condition: {day}\nNight condition: {night}\n")
                continue
                
            query_embeddings = model.encode(queries)
            closest_n = 2
            for query, query_embedding in zip(queries, query_embeddings):
                distances = scipy.spatial.distance.cdist([query_embedding], corpus_embeddings, "cosine")[0]

                results = zip(range(len(distances)), distances)
                results = sorted(results, key=lambda x: x[1])

                print("\n======================\n")
                if len(results) > 0:
                    idx, best_score = results[0]
                    if best_score > 0.16:
                        print("Proper answer not found from BERT, look DockDockGo instead:")
                        resp = search_dock(text)
                        if resp is None:
                            print("Sorry, Not good result found in Internet!")
                        else:
                            print(resp[:150]+"...")
                    else:
                        print("Query:", query)
                        print("\nTop 2 most similar sentences in corpus:")
                        for idx, distance in results[0:closest_n]:
                            print("Q: "+corpus[idx].strip(), "(Score: %.4f)" % (1-distance))
                            print("A: "+df2.iloc[idx,1].strip(), "(Score: %.4f)" % (1-distance))
            print("========================================================")
```
